```python
barWidth = 1

for col in data_used.columns:
    print(col)
    fig = plt.subplots(figsize =(8, 8))
    school_gone =  np.array(data_used[col])
    if(col == "States/UTs"):
        state =  np.array(data_used[col])
#     plt.scatter( school_gone,np.array(data_used["States/UTs"]))
    plt.bar(state,school_gone, color ='orange', width = barWidth, edgecolor ='grey')
    plt.xticks(rotation = 90)
    plt.show()
# plt.bar(br1,school_gone, color ='orange', width = barWidth, edgecolor ='grey',index = data_used["States/UTs"])
```

```python
num_clusters = 6
data_path = "nfhs.csv"
file = open(data_path)
csvreader = csv.reader(file)
header = []
header = next(csvreader)
rows = []
for row in csvreader:
    rows.append(row)
n = len(rows)
i = 5
dataset = []
state_names = []
while(i < n):
    dataset.append(rows[i][2:])
    state_names.append(rows[i][0])
    i = i + 3
for i in range(len(dataset)):
    for j in range(len(dataset[i])):
        if(dataset[i][j] == '*'):
            dataset[i][j] = 0
dataset = np.array(dataset)
kmeans = KMeans(n_clusters=num_clusters, random_state=0).fit(dataset)
labels = kmeans.labels_
output = {}
for i in range(num_clusters):
    output[i] = []
for i in range(len(labels)):
    output[labels[i]].append(state_names[i])
for i in range(num_clusters):
    print(output[i])
```

```python
for col in data_used.columns[2:]:
    print(col)
    for i in output.keys():
#         
        sum = 0
        for j in output[i]:
            state = data_used.loc[data_used['States/UTs'] == j]
            sum += float(np.array(state[col])[0])
        print(output[i])
        print(sum/len(output[i]))
```
